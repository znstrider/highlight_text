```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from highlight_text import HighlightText, ax_text, fig_text
```

```python
# Using matplotlibs new subplot_mosaic feature we can lay out our plot in advance.
layout = [['title'],
          ['main'],
          ['annot']]
# We can also do more complex layouts like the one shown here, see the docs here:
# https://matplotlib.org/3.3.0/tutorials/provisional/mosaic.html


# gridspec_kw allows us to set the height_ratios and width_ratios
fig, axd = plt.subplot_mosaic(layout,
                              constrained_layout=False,
                              figsize=(13, 10),
                              gridspec_kw={'height_ratios': [0.2, 0.6, 0.2]})

bbox_pad = 1.5
bboxprops = {'linewidth': 0, 'pad': bbox_pad}

# ax_text allows us to highlight text within delimiters <> using different
# colors, weights, bbox_kws and path_effect_kws for each <subtext>
title=\
ax_text(s='<Das Infektionsgeschehen in Europa>\n\n'
          'Tglich besttigte Neuinfektionen in <Tschechien>  <Belgien>  <Niederlande>  <Spanien>\n'
          '<Italien>  <Deutschland> und <weiteren Lndern> (7-Tage Durchschnitt)*',
        x=0,
        y=0.4,
        va='center',
        highlight_textprops=[{'fontsize': 24, 'color': 'k', 'weight': 'bold'},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#67001f', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#b2182b', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#fdae61', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#053061', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#4393c3', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#92c5de', **bboxprops}},
                             {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':'#969696', **bboxprops}}],
        ax=axd['title'],
        fontname='Roboto',
        fontsize = 16
       );


texts = ax_text(s='<*Aufgrund der deutlich hheren Zahl an Tests sind die Werte aus dem Frhjahr nur eingeschrnkt mit>\n'
                  '<denen aus dem Herbst vergleichbar>\n\n'
                  'Quelle: <Johns Hopkins CSSE>, Stand 14.10.2020, eigene Berechnung        Plotdesign: <DER SPIEGEL>',
                highlight_textprops=[{'color': 'k', 'weight': 'regular'}, 
                                     {'color': 'k', 'weight': 'regular'}, 
                                     {'color': 'k', 'weight': 'bold'},
                                     {'color': '#fc5203', 'weight': 'bold'}],
                color='#969696',
                ax=axd['annot'],
                x=0, y=0.1, fontsize=13, va='bottom')

axd['title'].axis('off');
axd['annot'].axis('off');
axd['main'].tick_params(axis='both', colors='None')

axd['main'].text(x=0.5, y=0.5, s='Main', color='darkgrey', fontsize=48, ha="center", va="center")

#plt.savefig('../examples/title_bboxes_example-spiegel.png', facecolor='w', dpi=200, bbox_inches='tight')
```
