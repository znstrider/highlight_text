```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from highlight_text import HighlightText, ax_text, fig_text
```

```python
# https://twitter.com/petermckeever/status/1346075580782047233

def identify_axes(ax_dict, fontsize=48):
    """
    Helper to identify the Axes in the examples below.
    Draws the label in a large font in the center of the Axes.
    
    SOURCE: https://matplotlib.org/stable/tutorials/provisional/mosaic.html

    Parameters
    ----------
    ax_dict : Dict[str, Axes]
        Mapping between the title / label and the Axes.

    fontsize : int, optional
        How big the label should be
    """
    kw = dict(ha="center", va="center", fontsize=fontsize, color="darkgrey")
    for k, ax in ax_dict.items():
        ax.text(0.5, 0.5, k, transform=ax.transAxes, **kw)
        
plt.style.use('dark')

main = np.arange(20).reshape(5, 4)

layout = np.concatenate([np.tile(['title'], (1, 4)),
                          main,
                         np.tile(['annot'], (1, 4))])

fig, axd = plt.subplot_mosaic(layout,
                              constrained_layout=True,
                              figsize=(12, 12),
                              gridspec_kw={'height_ratios': [0.1,  0.17, 0.17, 0.17, 0.17, 0.17, 0.05]})

# Here we use different heightlight_colors to inset the legend into the title
title=\
ax_text(s="How are the best creators in Europe's top 5 leagues making chances\n"
          "through <Corners>, <Crosses>, <Indirect Freekicks> and <Open Play Passes>?",
        x=0,
        y=0.5,
        va='center',
        color='w',
        fontweight='bold',
        highlight_textprops=[{'color': 'SkyBlue'}, 
                             {'color': 'orange'}, 
                             {'color': '#969696'},
                             {'color': 'red'}],
        ax=axd['title'],
        fontsize = 20
       )

for ax in axd.values():
    ax.tick_params(axis='both', colors='None')
    
identify_axes({'annotation axes - Plotdesign by Peter McKeever (@petermckeever)': axd['annot']}, fontsize=16)
    
axd['title'].axis('off');
axd['annot'].axis('off');

#plt.savefig('../examples/color_encoded_title-petermckeever.png', dpi=200, facecolor=fig.get_facecolor())
```
